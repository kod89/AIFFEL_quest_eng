## Bounding box 얼굴 검출

```python
#########이미지 파일 경로 저장 및 불러오기##########
home_dir = os.getenv('HOME')
my_image_path = os.path.join(home_dir, 'work/camera_sticker/images/2myimage.png') # 본인 이미지가 있는 경로를 가지고 온다.
img_bgr = cv2.imread(my_image_path)    # OpenCV로 이미지를 불러옵니다
img_show = img_bgr.copy()      # 출력용 이미지를 따로 보관
####################################################

# detector를 선언
detector_hog = dlib.get_frontal_face_detector() # 기본 얼굴 감지기를 반환
dlib_rects = detector_hog(img_show, 1)   # (image, num of image pyramid) 
# detector_hog의 두 번째 파라미터는 이미지 피라미드의 수
print(dlib_rects)
for dlib_rect in dlib_rects: # 찾은 얼굴 영역의 좌표
    l = dlib_rect.left() # 왼쪽
    t = dlib_rect.top() # 위쪽
    r = dlib_rect.right() # 오른쪽
    b = dlib_rect.bottom() # 아래쪽

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA) # 시작점의 좌표와 종료점 좌표로 직각 사각형을 그림

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

```python
# 모델 로드 #
model_path = os.path.join(home_dir, 'work/camera_sticker/models/shape_predictor_68_face_landmarks.dat')
    # 저장한 landmark 모델의 주소를 model_path 변수에 저장
landmark_predictor = dlib.shape_predictor(model_path)
    # dlib 라이브러리의 shape_predictor 함수를 이용하여 모델을 불러옴
    # landmark_predictor는 RGB이미지와 dlib.rectangle를 입력 받고 dlib.full_object_detection를 반환
    # dlib.rectangle: 내부를 예측하는 박스
    # dlib.full_object_detection: 각 구성 요소의 위치와, 이미지 상의 객체의 위치를 나타냄
```

## Landmark 검출하기

```python
list_landmarks = []
    # 랜드마크의 위치를 저장할 list 생성

# 얼굴 영역 박스 마다 face landmark를 찾아냅니다
# face landmark 좌표를 저장해둡니다
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_show_rgb, dlib_rect)
        # 모든 landmark의 위치정보를 points 변수에 저장
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
        # 각각의 landmark 위치정보를 (x,y) 형태로 변환하여 list_points 리스트로 저장
    list_landmarks.append(list_points)
        # list_landmarks에 랜드마크 리스트를 저장

print(len(list_landmarks[0]))
    # 얼굴이 n개인 경우 list_landmarks는 n개의 원소를 갖고
    # 각 원소는 68개의 랜드마크 위치가 나열된 list
    # list_landmarks의 원소가 1개이므로 list_landmarks[1]을 호출하면 IndexError가 발생
```

```python
# list_landmarks의 원소가 1개이므로 아래 반복문은 한번만 실행됨
print(dlib_rects)
for landmark in list_landmarks:
    for point in landmark:
        cv2.circle(img_show_rgb, point, 2, (0, 255, 255), -1)
            # cv2.circle: OpenCV의 원을 그리는 함수
            # img_show 이미지 위 각각의 point에
            # 크기가 2이고 (0, 255, 255)색으로 내부가 채워진(-1) 원을 그림
            # (마지막 인수가 자연수라면 그만큼의 두께의 선으로 원이 그려짐)

    # RGB 이미지로 전환
plt.imshow(img_show_rgb)
    # 이미지를 준비
plt.show()
    # 이미지를 출력
```

```python
# zip() : 두 그룹의 데이터를 서로 엮어주는 파이썬의 내장 함수
# dlib_rects와 list_landmarks 데이터를 엮어 주었음
# dlib_rects : 얼굴 영역을 저장하고 있는 값
# → rectangles[[(32, 80) (139, 187)]]
# list_landmarks : 68개의 랜드마크 값 저장(이목구비 위치(x,y))


for dlib_rect, landmark in zip(dlib_rects, list_landmarks): # 얼굴 영역을 저장하고 있는 값과 68개의 랜드마크를 저장하고 있는 값으로 반복문 실행
    print (landmark[30]) # 코 index는 30
    x = landmark[30][0] # 이미지에서 코 부위의 x값
    y = landmark[30][1] # 이미지에서 코 부위의 y값
    mouth_left = landmark[48]
    mouth_right = landmark[54]
    w = int((mouth_right[0] - mouth_left[0]) * 2.0)
    h = int(w * img_sticker.shape[0] / img_sticker.shape[1])
    
    print (f'(x,y) : ({x},{y})')
    print (f'(w,h) : ({w},{h})')
```

```python
sticker_path = os.path.join(home_dir, 'work/camera_sticker/images/cat-whiskers.png') # 고양이 수염 이미지의 경로
img_sticker = cv2.imread(sticker_path, cv2.IMREAD_UNCHANGED) 
img_sticker = cv2.resize(img_sticker, (w,h))
print (img_sticker.shape)
```

```python
###### x,y,w,h 모두 위에서 반복문 안에서 지정해준 값임
refined_x = x - w // 2
refined_y = y - h // 2
print (f'(x,y) : ({refined_x},{refined_y})') 
```

```python
# if refined_x < 0:
#     img_sticker = img_sticker[:, -refined_x:]
#     refined_x = 0

# if refined_y < 0:
#     img_sticker = img_sticker[-refined_y:, :]
#     refined_y = 0

# print (f'(x,y) : ({refined_x},{refined_y})')
```

```python
sticker_area = img_show_rgb[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
##### np.where용 “마스크” 만들기 #####
alpha = img_sticker[:, :, 3]   # 0~255
mask = alpha > 0               # True = 수염, False = 투명
######################################
img_show_rgb[
    refined_y:refined_y+img_sticker.shape[0],
    refined_x:refined_x+img_sticker.shape[1]
] = np.where(
    mask[:, :, np.newaxis],
    img_sticker[:, :, :3],   # 수염 픽셀
    sticker_area             # 원본 이미지
).astype(np.uint8)

# img_show_rgb[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
#     np.where(img_sticker==0,sticker_area,img_sticker).astype(np.uint8)
```

### ❌ 문제 1
img_sticker = cv2.imread(sticker_path)


👉 투명 PNG인데 알파 채널을 안 읽음

이 상태: img_sticker.shape == (h, w, 3)

투명 정보(α)가 아예 사라짐

그래서 np.where(img_sticker==0, ...) 같은 꼼수 마스킹을 쓰게 됨  -> 무슨 의미인지 후에 시간되면 찾아보기

이게 하얀 영역 + 잔상 문제의 원인

✅ 해결 전략

img_sticker = cv2.imread(sticker_path) -> img_sticker = cv2.imread(sticker_path, cv2.IMREAD_UNCHANGED) 
* img_sticker.shape == (h, w, 4)

### ❌ 문제 2
w = h = dlib_rect.width()


👉 수염을 얼굴 전체 크기 정사각형으로 키움

수염은 가로로 길고 세로는 얇아야 정상

이 때문에 이마까지 덮음

✅ 해결 전략

w = h = dlib_rect.width()

-> 
mouth_left = landmark[48]
mouth_right = landmark[54]

w = int((mouth_right[0] - mouth_left[0]) * 2.0)
h = int(w * img_sticker.shape[0] / img_sticker.shape[1])


👉가로: 입 폭 × 2   /  세로: 이미지 비율 유지

### ❌ 문제 3
refined_y = y - h


👉 수염을 얼굴 높이만큼 위로 이동

코 기준이 아니라 이마 기준이 됨

✅ 해결 전략

refined_x = x - w // 2 
refined_y = y - h

->
refined_x = x - w // 2
refined_y = y - h // 2

👉 이제 수염 중앙이 코에 정확히 위치

#### [np.where가 왜 문제가 되었는지 짚고 가기] (by.GPT)

예시 코드의 핵심 줄:

- np.where(img_sticker==0, sticker_area, img_sticker)


이건 이런 의미

: img_sticker의 RGB 값이 정확히 0(검정) 인 픽셀만 투명 취급

문제:

투명 PNG의 “투명함”은 알파 채널(α) 에 있음

RGB 값이 0이냐 아니냐와 전혀 무관

그래서

흰 배경이 남고

이전 흔적이 남고

경계가 이상해짐

👉 np.where 자체가 문제는 아님

👉 조건식이 잘못된 것

np.where를 쓰려면 반드시 “마스크”가 필요함

np.where의 정석 구조:

np.where(조건, A, B)


조건이 True → A

조건이 False → B

### 원본에 반영

```python
sticker_area = img_bgr[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]

alpha = img_sticker[:, :, 3]   
mask = alpha > 0               
img_bgr[
    refined_y:refined_y+img_sticker.shape[0],
    refined_x:refined_x+img_sticker.shape[1]
] = np.where(
    mask[:, :, np.newaxis],
    img_sticker[:, :, :3],   
    sticker_area             
).astype(np.uint8)
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.show()
```
