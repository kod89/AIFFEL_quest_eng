```python
import random as r
class Account:
    
    total_account = 0
    
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        self.bank_name = 'SC은행'
        self.deposit_count = 0
        self.deposit_list = []
        self.withdraw_list = []
        num_1 = str(r.randint(0, 999)).zfill(3)    #zfill(자릿수) : 앞에 빈 숫자 0 끼워넣어 원하는 자릿수 만들어줌
        num_2 = str(r.randint(0, 99)).zfill(2)
        num_3 = str(r.randint(0, 999999)).zfill(6)
        account_number = num_1 + '-' + num_2 + '-' + num_3
        self.account_number = account_number
        Account.total_account += 1
    
    @classmethod                # 클래스메서드 
    def get_account_num(cls):   # cls는 인스턴스 메서드의 self의 기능
        print(f"개설된 계좌의 총 개수는 {cls.total_account}개 입니다")

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            self.deposit_list.append(amount)
            self.deposit_count += 1
            print(f"입금이 완료 되었습니다. 잔액 : {self.balance} 원")
            if self.deposit_count % 5 == 0:
                self.balance = int(self.balance * 1.01)
                print(f"이자가 지급되었습니다. 잔액 : {self.balance} 원")
        else:
            print("1원 이상부터 입금이 가능합니다.")

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            self.withdraw_list.append(amount)
            print(f"출금이 완료 되었습니다. 잔액 : {self.balance} 원")
        else:
            print(f"잔고 이상으로 출금할 수 없습니다. 현재 잔액 : {self.balance} 원")

    def display_info(self):
        print(f"은행이름 : {self.bank_name}")
        print(f"예금주 : {self.name}")
        print(f"계좌번호 : {self.account_number}")
        print(f"잔고 : {self.balance:,} 원")             # f"{숫자:,}" - 세자리 마다 쉼표

    def deposit_history(self):
        for amount in self.deposit_list:
            print(f"{amount}원 입금")

    def withdraw_history(self):
        for amount in self.withdraw_list:
            print(f"{amount}원 출금")
            
```

```python
a = Account("대홍1",1000)
print(a.name)
print(a.balance)
print(a.bank_name)
print (a.account_number)
print(Account.total_account)
```

```python
b = Account("대홍2",50000)
c = Account("대홍3",2000000)
```

```python
a.deposit(3000)
```

```python
a.withdraw(1000)
```

```python
a.display_info()
```

```python
clients = [a, b, c]
```

```python
for client in clients:
    if client.balance >= 1000000:
        client.display_info()
```

```python
a.deposit_history()
```

```python
a.withdraw_history()
```
