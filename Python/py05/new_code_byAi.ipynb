```python
# AI가 작성한 코드  #

import random as r

class Account:
    bank_name = "SC은행"      # 모든 계좌가 공통으로 쓰는 값 → 클래스 변수
    total_account = 0         # 생성된 계좌 수

    def __init__(self, name, balance):
        self.name = name
        self.balance = int(balance)  # 잔고는 정수로 유지(원 단위)
        self.account_number = self._make_account_number()

        # Q7: 입금 횟수
        self.deposit_count = 0

        # Q10: 입금/출금 내역
        self.deposit_list = []
        self.withdraw_list = []

        Account.total_account += 1

    # 계좌번호 생성 로직을 분리(가독성 ↑)
    def _make_account_number(self):
        num_1 = str(r.randint(0, 999)).zfill(3)
        num_2 = str(r.randint(0, 99)).zfill(2)
        num_3 = str(r.randint(0, 999999)).zfill(6)
        return f"{num_1}-{num_2}-{num_3}"

    @classmethod
    def get_account_num(cls):
        print(f"개설된 계좌의 총 개수는 {cls.total_account}개 입니다")

    def deposit(self, amount):
        amount = int(amount)

        if amount <= 0:
            print("1원 이상부터 입금이 가능합니다.")
            return

        self.balance += amount
        self.deposit_list.append(amount)
        self.deposit_count += 1

        # Q7: 5회(5의 배수)마다 1% 이자 지급
        if self.deposit_count % 5 == 0:
            interest = int(self.balance * 0.01)
            self.balance += interest
            print(f"이자가 지급되었습니다. 이자: {interest:,}원, 잔액: {self.balance:,}원")
        else:
            print(f"입금이 완료 되었습니다. 잔액 : {self.balance:,} 원")

    def withdraw(self, amount):
        amount = int(amount)

        if amount <= 0:
            print("1원 이상부터 출금이 가능합니다.")
            return

        if self.balance < amount:
            print(f"잔고 이상으로 출금할 수 없습니다. 현재 잔액 : {self.balance:,} 원")
            return

        self.balance -= amount
        self.withdraw_list.append(amount)
        print(f"출금이 완료 되었습니다. 잔액 : {self.balance:,} 원")

    def display_info(self):
        print(f"은행이름 : {Account.bank_name}")
        print(f"예금주 : {self.name}")
        print(f"계좌번호 : {self.account_number}")
        print(f"잔고 : {self.balance:,} 원")

    def deposit_history(self):
        for amount in self.deposit_list:
            print(f"{amount:,}원 입금")

    def withdraw_history(self):
        for amount in self.withdraw_list:
            print(f"{amount:,}원 출금")
```

```python
# 공통값 bank_name을 클래스 변수로 올림 (Account.bank_name) → 중복 저장 줄임

# 계좌번호 생성 로직을 _make_account_number()로 분리 → __init__ 가독성 ↑

# 이자 지급을 self.balance = int(self.balance * 1.01) 대신
# interest = int(self.balance * 0.01); self.balance += interest로 → 이자 금액을 명확히 기록/출력 가능

# deposit/withdraw에서 amount <= 0 처리 추가 → 입력이 이상할 때 방어적

# 출력에서 잔액/금액에 :, 적용 → Q6 스타일 일관
```
